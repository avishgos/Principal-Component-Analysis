About the data set  (Life Expectancy data)Â¶
The dataset contains information about several health and economic factors that contribute to average life expectancy for different countries.
Attribute information:

Country: Name of the country

Status: Whether the country is Developed or Developing

Adult_Mortality: Mortality rate for age group 15-60 out of every 1000 individuals of the population

Infant_Deaths: Number of infant deaths per 1000 population

Hepatitis: Hepatitis B (HepB) immunization coverage for 1 year olds (Percentage)

Measles: Number of reported cases for measles per 1000 from population

BMI: Average Body Mass Index for entire population

Underfive_Deaths: Number of deaths under 5 years of age per 1000 population

Polio: Polio (Pol3) immunization coverage for 1 year olds (Percentage)

Diphtheria: Diphtheria tetanus toxoid and pertussis (DTP3) immunization coverage for 1 year olds (Percentage)

HIV: Deaths per 1000 live births due to HIV/AIDS (0-4 years)

GDP: Gross Domestic Product per capita (in USD)

Population: Population of the country

Malnourished10_19: Prevalence of malnutrition among children and adolescents for Age 10 to 19 (Percentage)

Malnourished5_9: Prevalence of malnutrition among children for Age 5 to 9 (Percentage)

Income_Index: Human Development Index (HDI) in terms of national income per capita (index ranging from 0 to 1)

Schooling: Number of years of Schooling

Life_Expectancy: Life Expectancy in age for the country

```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing dataset and seeing the head of df
df = pd.read_csv('Life_Expectancy.csv')
df.head()
```

```python
#Lets see the shape of df(rows,columns)
df.shape
```

```python
#Dropping country as its unique in nature
df = df.drop('Country',axis = 1)
df.head()
```

```python
#Checking data types
df.info()
```

```python
#Lets see the five summary statistics
df.describe()
```

```python
cor=df.corr()
# plotting correlations on a heatmap

# figure size
plt.figure(figsize=(20,10))

# heatmap
sns.heatmap(cor, cmap="YlGnBu", annot=True)
plt.show()
```

```python
#encoding Status
df['Status'] = df['Status'].map({"Developing":0,'Developed':1})
```

```python
#value count of status
df['Status'].value_counts()
```

```python
#importing zscore
from scipy.stats import zscore
```

```python
#Appyling zscore and standardizing it
df = df.apply(zscore)
df.head()
```

```python
#Building kmeans model and checking the best K-value

from sklearn.cluster import KMeans


wcss = [] #inertia (within cluster sum square) or total spread

for k in range(1,10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
    
# Visualization of k values:

plt.plot(range(1,10), wcss, color='red',marker = 'o')
plt.title('Graph of k values and WCSS')
plt.xlabel('k values')
plt.ylabel('wcss values')
plt.show()
```

- At K = 2 ,error is reduced at large,after that there is constant(Small) reduction

```python
#Displaying all the k values
wcss
```

```python
#now we know our best k value is 2 so building model with k = 2

from sklearn.cluster import KMeans 
kmeans = KMeans(n_clusters = 2,n_init = 10,random_state = 100)

#training the model
clusters = kmeans.fit_predict(df)

#adding a label feature with predicted class value
df_k = df.copy(deep = True)
df_k['labels'] = clusters
df_k.head()
```

```python
#Calculating inertia
kmeans.inertia_
```

```python
#Calculating silhouette score to checl the overall quality of cluster
from sklearn import metrics

#used to judge the overall quality of Cluster
metrics.silhouette_score(df,kmeans.labels_)
```

```python
#value count for cluster
df_k['labels'].value_counts()
```

```python
#Converting the value count from 1,0 to 0,1
df_k['labels'] = 1-df_k['labels']
```

```python
# Again checking value count for cluster
df_k['labels'].value_counts()
```

```python
#Check centroid value for each cluster and store it in new dataframe
centroids = kmeans.cluster_centers_
centroids_df = pd.DataFrame(centroids,columns = list(df))
centroids_df
```

# Agglomerative clustering for Kmeans

```python
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage
from scipy.spatial.distance import pdist  #Pairwise distribution between data points
# cophenet index is a measure of the correlation 
#between the distance of points in feature space and 
#distance on dendrogram
# closer it is to 1, the better is the clustering
#min or single | max or complete | 
```

```python
Z = linkage(df, 'complete')   # cophenetic distance (single/complete/average)
c, coph_dists = cophenet(Z , pdist(df))

c
```

```python
Z1 = linkage(df, 'single')   # cophenetic distance (single/complete/average)
c, coph_dists = cophenet(Z1 , pdist(df))

c
```

```python
Z2 = linkage(df, 'average')   # cophenetic distance (single/complete/average)
c, coph_dists = cophenet(Z2 , pdist(df))

c
```

```python
Z3 = linkage(df, 'ward')   # cophenetic distance (single/complete/average)
c, coph_dists = cophenet(Z3 , pdist(df))

c
```

- **Average is the best linkage method with 0.88**
- **Closer it is to 1 better it is**

## Applying PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 17)  #eigen vectors
principalComponents = pca.fit_transform(df)
pricipalDF = pd.DataFrame(data = principalComponents)
```

- How each of the principle component is contributing to the mode

```python
pricipalDF.shape
```

```python
explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
cum_var_exp = np.cumsum(explained_variance)  # an array of cumulative variance. There will be 8 entries with 8 th entry 
# cumulative reaching almost 100%
print(cum_var_exp)
```

- We can see that after 12 values are constatnt so we wont take those values

```python
plt.bar(range(1,18), explained_variance, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(1,18),cum_var_exp, where= 'mid', label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc = 'best')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=12)
principalComponents = pca.fit_transform(df)
principalDf = pd.DataFrame(data = principalComponents,columns = list(['PC0','PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11']))
explained_variance = pca.explained_variance_ratio_ 
print(explained_variance)
cum_var_exp = np.cumsum(explained_variance)  # an array of cumulative variance. There will be 8 entries with 8 th entry 
# cumulative reaching almost 100%
print(cum_var_exp)
```
